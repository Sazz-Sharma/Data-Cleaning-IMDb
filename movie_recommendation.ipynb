```python
def change_to_nan(x):
    if x == 'Not Rated' or x =='Not Available' or x=='Unrated':
        return np.nan
    else:
        return x
```

```python
changed_certificate = df['certificate'].apply(change_to_nan, convert_dtype=True)
```

```python
df = df.drop(['certificate'], axis = 1)
```

```python
df.loc[:, "certificate"] = changed_certificate
```

```python
df['box_office'] = df['box_office'].apply(change_to_nan)
```

```python
df['budget'] = df['budget'].apply(change_to_nan)
```

```python
df = df.applymap(change_to_nan)
```
